```python
#dependiences
import pandas as pd
import datetime as dt
```

```python
#read the csv
pd.options.display.max_colwidth = 200
best_seller_df = pd.read_csv("bestsellers.csv")
best_seller_df
```

```python
#check datatypes
best_seller_df.dtypes
```

```python
#clean up the possible nulls
best_seller_df =best_seller_df.dropna()
best_seller_df
```

```python
#drop unnessary data
best_seller_df = best_seller_df.drop(columns = ["isbn10", "isbn13", "amazon_product_url","list_name" ])
best_seller_df
```

```python
#check datatypes for cleaning
best_seller_df.dtypes
```

```python
#change published_date to datetime
best_seller_df["published_date"] =  pd.to_datetime(best_seller_df["published_date"])
best_seller_df
```

```python
#check if the datatype was changed
best_seller_df.dtypes
```

```python
#replace the '-' with '_' in the list_name_encoded
best_seller_df["list_name_encoded"] = best_seller_df["list_name_encoded"].str.replace('-', '_')
best_seller_df.head(10)
```

```python
#rename list_name_encoded to book_type
best_seller_df = best_seller_df.rename(columns={'list_name_encoded':'book_type'})
best_seller_df.head()
```

```python
#clean up the list_name_encoded
best_seller_df.to_csv("cleaned_best_seller_code.csv", index=False)
```
